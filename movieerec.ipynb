```
arr=movies.iloc[0:,2]
```

```
genre=arr.str.get_dummies(sep='|')
```

```
genre['movieId']=movies['movieId']
```

```
mod_movies=pd.merge(movies, genre, how ='inner',on='movieId')
mod_movies
```

```
from sklearn.model_selection import train_test_split
```

```
#split into training and testing data 
train_data,test_data=train_test_split(ratings,test_size=0.2)
```

```
Data_rat=pd.merge(ratings,movies,on='movieId')
```

```
Data_rat.head()
```

```
#most rated movies ---top 25
most_rated=Data_rat.groupby('title').size().sort_values(ascending=False)[:25]
most_rated
```

```
import numpy as np
```

```
#most highly rated movies

#grouping movieId with mean rating
movie_stats=Data_rat.groupby('title').agg({'rating': [np.size, np.mean]})
movie_stats.head()
```

```
# sort by rating average
movie_stats.sort_values([('rating', 'mean')], ascending=False).head()
```

```
#most higly rated movies with atleast 100 ratings
atleast = movie_stats['rating']['size'] >= 100
movie_stats[atleast].sort_values([('rating', 'mean')], ascending=False)[:10]
```
